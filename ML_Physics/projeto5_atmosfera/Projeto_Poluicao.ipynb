```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing


# Fit the model
model_ets = ExponentialSmoothing(dfs['O3'],
                                 trend='add',
                                 seasonal='add',  # Additive seasonality
                                 seasonal_periods=12)  # Monthly seasonality
fitted_model_ets = model_ets.fit()

# Print model parameters
print(fitted_model_ets.params)

# Residuals analysis
residuals_ets = pd.DataFrame(fitted_model_ets.resid)
residuals_ets.plot(title='Gráfico dos resíduos')
plt.show()

# Density plot of residuals
residuals_ets.plot(kind='kde', title='Distribuição dos resíduos')
plt.show()

# Summary stats of residuals
print(residuals_ets.describe())

# Aplicando o modelo aos dados
pred_ets = fitted_model_ets.fittedvalues
plt.plot(dfs['O3'].index, dfs['O3'], label='Observado')
plt.plot(pred_ets.index, pred_ets, label='Predito')
plt.ylabel('Concentração (O3)')
plt.title('Modelo de Suavização Exponencial')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller


# Check for stationarity and difference the data if necessary
def check_stationarity(series, signif=0.05):
    """ Perform ADFuller to test for stationarity """
    result = adfuller(series, autolag='AIC')
    p_value = result[1]
    return p_value < signif

# Differencing non-stationary series
df_stationary = dfs.copy()
for col in dfs.columns:
    if not check_stationarity(dfs[col]):
        df_stationary[col] = dfs[col].diff().dropna()

# Drop NaN values created by differencing
df_stationary = df_stationary.dropna()

# Fit the VAR model
model_var = VAR(df_stationary)
fitted_model_var = model_var.fit(maxlags=15, ic='aic')

# Print the summary of the fitted model
print(fitted_model_var.summary())

# Residuals analysis
residuals_var = pd.DataFrame(fitted_model_var.resid)
residuals_var['O3'].plot(title='Gráfico dos resíduos')
plt.show()

# Density plot of residuals
residuals_var['O3'].plot(kind='kde', title='Distribuição dos resíduos')
plt.show()

# Summary stats of residuals
print(residuals_var['O3'].describe())

# Forecasting
lag_order = fitted_model_var.k_ar
forecast_input = df_stationary.values[-lag_order:]
forecast = fitted_model_var.forecast(y=forecast_input, steps=10)
forecast_df = pd.DataFrame(forecast, index=pd.date_range(start=dfs.index[-1], periods=10, freq='M'), columns=dfs.columns)

# Extract the forecast for O3
forecast_o3 = forecast_df['O3']

# Aplicando o modelo aos dados
plt.plot(dfs['O3'].index, dfs['O3'], label='Observado')
plt.plot(forecast_o3.index, forecast_o3, label='Predito')
plt.ylabel('Concentração (O3)')
plt.title('Modelo VAR para O3')
plt.legend()
plt.show()
```
